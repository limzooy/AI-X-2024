```python
#Q1) 2019.01~06중에 언제 지하철을 가장 많이 이용했을까? (기준: 승하차총승객수)
#Q1-1 가장 많이 이용한 월 -> 월별로 그룹화, sum

# 사용일자 열을 날짜 형식으로 변환
df['사용일자'] = pd.to_datetime(df['사용일자'])

# 월별로 그룹화하여 승하차 총승객수의 합을 계산
monthly_usage = df.resample('M', on='사용일자').sum()

# 가장 많은 이용량을 기록한 달 찾기
max_month = monthly_usage['승하차총승객수'].idxmax().strftime('%Y-%m')
max_usage = monthly_usage['승하차총승객수'].max()

print(f"{max_month}에 가장 많은 이용량을 기록했습니다. 이용량: {max_usage}") # <- GPT사용3
```

```python
#Q1-2 가장 많이 이용한 일 -> 사용일자로 그룹화 + 합

# 날짜별로 그룹화하여 승하차 총승객수의 합을 계산
daily_usage = df.groupby('사용일자').sum()

# 가장 많이 이용한 날 찾기
max_usage_date = daily_usage['승하차총승객수'].idxmax().strftime('%Y-%m-%d')
max_usage_count = daily_usage['승하차총승객수'].max()

print(f"{max_usage_date}에 가장 많이 이용되었습니다. 이용량: {max_usage_count}")
```

```python
#Q3 가설) 요일중에서 목요일에 지하철 승객수가 많다? (기준: 승하차총승객수)
# -> 요일별로 승하차총승객수를 보자

# 사용일자 열을 날짜 형식으로 변환
df['사용일자'] = pd.to_datetime(df['사용일자'])

# 요일별로 그룹화하여 승하차 총승객수의 합 계산
daily_usage = df.groupby(df['사용일자'].dt.dayofweek)['승하차총승객수'].sum()

# 요일 이름 변경
weekday_names = ['일', '월', '화', '수', '목', '금', '토']
daily_usage.index = weekday_names

# 시각화
plt.figure(figsize=(10, 6))
daily_usage.plot(kind='bar', color='skyblue')
plt.title('요일별 지하철 승하차 총승객수')
plt.xlabel('요일')
plt.ylabel('총승객수')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 가설 결과: 목요일이 아닌 금요일이 총 승객수가 제일 많다. 
```

```python
#Q4) 연월 각각에 대해 일자별(월일별) 승하차총승객수 그래프를 그려보세요. (pointplot)

# 사용일자 열을 날짜 형식으로 변환
df['사용일자'] = pd.to_datetime(df['사용일자'])

# 연월 열 추가
df['연월'] = df['사용일자'].dt.strftime('%Y-%m')

# 연월과 월일 열을 합쳐서 새로운 열 생성
df['연월일'] = df['연월'] + '-' + df['월일']

# 연월별로 그룹화하여 일별 승하차 총승객수의 합 계산
daily_usage = df.groupby(['연월', '연월일'])['승하차총승객수'].sum().reset_index()

# 시각화
plt.figure(figsize=(14, 8))
sns.pointplot(data=daily_usage, x='연월일', y='승하차총승객수', hue='연월', dodge=True, markers='.', ci=None)
plt.title('연월별 일별 지하철 승하차 총승객수')
plt.xlabel('일자')
plt.ylabel('승하차총승객수')
plt.xticks(rotation=45, ha='right')
plt.legend(title='연월', bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 사용일자 열을 날짜 형식으로 변환
df['사용일자'] = pd.to_datetime(df['사용일자'])

# 연월 열 추가
df['연월'] = df['사용일자'].dt.strftime('%Y-%m')

# 연월별로 그룹화하여 일별 승하차 총승객수의 합 계산
daily_usage = df.groupby(['연월', '월일'])['승하차총승객수'].sum().reset_index()

# 월별로 서브플롯에 그래프 그리기
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# 각 월별로 그래프 그리기
for i, (month, data) in enumerate(daily_usage.groupby('연월')):
    row = i // 3
    col = i % 3
    sns.pointplot(data=data, x='월일', y='승하차총승객수', ax=axes[row, col], markers='.', ci=None)
    axes[row, col].set_title(month)
    axes[row, col].set_xlabel('일자')
    axes[row, col].set_ylabel('총승객수')
    axes[row, col].set_xticklabels(axes[row, col].get_xticklabels(), rotation=45, ha='right')

# 서브플롯 간 간격 조정
plt.tight_layout()
plt.show()
```

```python
#Q5) 가장 승객이 많이 타는 승차역은?
#승차총승객수가 가장 많은 곳??
#Q5-1) 1일 기준

# 승차총승객수가 가장 큰 행 찾기
most_boarded_station = df.loc[df['승차총승객수'].idxmax()]

# 결과 출력
print("가장 승객이 많이 타는 승차역:")
print(most_boarded_station)
```

```python
#Q5-1) 누적으로 승객이 가장 많이 탄 승차역
# 그럼 역명별로 승차총객수를 합쳐야함.


# 역별로 승차총승객수 합산
station_boarding_total = df.groupby('역명')['승차총승객수'].sum()

# 가장 승객이 많이 타는 역 찾기
most_boarded_station = station_boarding_total.idxmax()
max_boarding_count = station_boarding_total.max()

# 결과 출력
print(f"가장 승객이 많이 타는 승차역: {most_boarded_station}")
print(f"누적 승차총승객수: {max_boarding_count}")
```

```python
#Q6) 노선별로 역별/요일별 승차승객수를 비교해 볼 수 있을까? (1~9호선, 역별/요일별 heatmap)
# 노선정보, 역명, 요일, 승차총승객수 필요


# 필요한 열만 선택 (노선명, 역명, 요일, 승차총승객수)
df1 = df[['노선명', '역명', '요일', '승차총승객수']]

# 노선별로 그룹화
grouped = df1.groupby('노선명')

# 노선별로 역별/요일별 승차승객수 합계 계산 및 시각화
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 14))

for i, (name, group) in enumerate(grouped):
    row = i // 3
    col = i % 3
    
    pivot_table = group.pivot_table(index='역명', columns='요일', values='승차총승객수', aggfunc='sum')
    
    sns.heatmap(pivot_table, cmap='Blues', ax=axes[row, col], cbar=False)
    axes[row, col].set_title(name)
    axes[row, col].set_xlabel('요일')
    axes[row, col].set_ylabel('역명')

plt.tight_layout()
plt.show()
```

```python
#Q7) 1호선에서 가장 하차를 많이 하는 역은? (groupby)

# 필요한 열만 선택 (노선명, 역명, 하차총승객수)
df2 = df[['노선명', '역명', '하차총승객수']]

# 1호선 데이터만 필터링
line_1_data = df2[df2['노선명'] == '1호선']

# 역별 하차총승객수 합계 계산
station_exit_total = line_1_data.groupby('역명')['하차총승객수'].sum()

# 가장 하차를 많이 하는 역 찾기
most_exited_station = station_exit_total.idxmax()
max_exit_count = station_exit_total.max()

# 결과 출력
print(f"1호선에서 가장 하차를 많이 하는 역: {most_exited_station}")
print(f"하차총승객수: {max_exit_count}")
```
